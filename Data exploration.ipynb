```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("Zara_Sales_Analysis.csv", delimiter=';')

# Basic information
print("Shape:", df.shape)
print("\nInfo:")
print(df.info())
print("\nMissing values:\n", df.isnull().sum())

# Display first few rows
print("\nFirst 5 rows:\n", df.head())

# Summary statistics for numerical columns
print("\nDescriptive statistics:\n", df.describe())

# Check unique values for categorical columns
categorical_cols = ['Product Position', 'Promotion', 'Product Category', 'Seasonal', 'section']
for col in categorical_cols:
    print(f"\nUnique values in '{col}': {df[col].unique()}")

# Missing value heatmap
plt.figure(figsize=(10, 4))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()

# Distribution of Sales Volume
plt.figure(figsize=(8, 4))
sns.histplot(df['Sales Volume'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Sales Volume")
plt.xlabel("Sales Volume")
plt.ylabel("Frequency")
plt.show()

# Boxplot of Sales Volume by Promotion
plt.figure(figsize=(6, 4))
sns.boxplot(x='Promotion', y='Sales Volume', data=df)
plt.title("Sales Volume by Promotion")
plt.show()

# Boxplot of Sales Volume by Seasonal
plt.figure(figsize=(6, 4))
sns.boxplot(x='Seasonal', y='Sales Volume', data=df)
plt.title("Sales Volume by Seasonal Products")
plt.show()

# Correlation heatmap for numeric variables
numeric_cols = ['Sales Volume', 'price']
plt.figure(figsize=(6, 4))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# Count plots for categorical variables
for col in categorical_cols:
    plt.figure(figsize=(6, 3))
    sns.countplot(data=df, x=col)
    plt.title(f"Count of {col}")
    plt.xticks(rotation=45)
    plt.show()

# Price vs Sales Volume
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='price', y='Sales Volume', hue='Promotion')
plt.title("Price vs Sales Volume")
plt.show()
```
